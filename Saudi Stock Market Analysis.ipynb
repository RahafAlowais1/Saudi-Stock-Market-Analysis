## Data Clening

```python
data['open']= data['open'].fillna(np.mean(data['open']))
data['high']= data['high'].fillna(np.mean(data['high']))
data['low']= data['low'].fillna(np.mean(data['low']))
```

```python
data[data.duplicated()]
```

```python
data.columns
```

```python
data = data.rename (columns= {'sectoer' : 'sector' ,'trading_name ' :'trading_name'  , 'volume_traded ': 'volume_traded','no_trades ' :'no_trades'}) # تعديل اسماء الكولم

data. columns
```

```python
data ['date'] = pd.to_datetime(data['date'])

data.dtypes
```

```python
data ['no_trades'] = data['no_trades'].astype('int')
data .dtypes
```

## Data Analysis & Visualization

```python
plt.figure(figsize=(8, 4))

plt.plot(data['close'], label=' closing Price', color='r')# Set the figure size
plt.plot(data['open'], label='opening Price', color='b')
plt.title("Line Plot:Closing price over Index")  # Add a title
plt.xlabel("Index")  # Label the x-axis
plt.ylabel("Closing price")  # Label the y-axis
plt.legend()  # Add a legend
plt.grid(True)  # Add a grid
plt.show()  # Display the plot
```

```python
data_sarco = data[ data['trading_name'] =='SARCO']

plt.figure(figsize=(14, 4))  # Set the figure size
plt.plot(data_sarco['date'], data_sarco['close'], label='closing Price', color='r' ,linestyle='solid',marker ='o')
plt.plot(data_sarco['date'], data_sarco['open'], label='opening Price', color='b' ,linestyle='solid',marker ='o')
plt.plot(data_sarco['date'], data_sarco['high'], label='high Price', color='g' ,linestyle='solid',marker ='o')
plt.plot(data_sarco['date'], data_sarco['low'], label='low Price', color='black' ,linestyle='solid',marker ='o')

plt.title("Line Plot: Closing price, opening price, high price and low price over time ")  # Add a title
plt.xlabel("Time")  # Label the x-axis
plt.ylabel("Closing price")  # Label the y-axis
plt.legend()  # Add a legend
plt.grid(True)  # Add a grid
plt.show()  # Display the plot
```

```python
plt.figure(figsize=(8, 5))
plt.hist(data['change'], bins=30, color='c', edgecolor='black')
plt.title("Histogram:Change Price")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
data_filtere = data [(data['change']<5) & (data['change']>-5)]

plt.figure(figsize=(8, 5))
plt.hist(data_filtere['change'], bins=30, color='c', edgecolor='black')
plt.title("Histogram: Change price")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
data_filtere = data [(data['perc_Change']<10) & (data['perc_Change']>-10)]

plt.figure(figsize=(8, 5))
plt.hist(data_filtere['perc_Change'], bins=30, color='c', edgecolor='black')
plt.title("Histogram: perc_Change")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.boxplot(data["change"], patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Box Plot: Change Price")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
data_filtered = data[(data['change'] < 1) & (data['change'] > -1)]

plt.figure(figsize=(8, 5))
plt.boxplot(data_filtered["change"], patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Box Plot: Change Price ")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

* Which sector has the highest average closing price?

```python
grouped_data= data.groupby(['sector'])['close'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
grouped_data.plot.bar()
plt.title("Mean Closing Price by Industry")
plt.xlabel("Industry")
plt.ylabel("Mean Closing Price")
plt.show()
```

- Are the sectors with the highest number of trades the same as those with the highest value traded?

```python
grouped_data= data.groupby(['sector'])['no_trades'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
grouped_data.plot.bar()
plt.title("Total number of Trades by Industry")
plt.xlabel("Industry")
plt.ylabel("total number of Trades ")
plt.show()
```

```python
grouped_data= data.groupby(['sector'])['value_traded'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
grouped_data.plot.bar()
plt.title("Total value of Trades by Industry")
plt.xlabel("Industry")
plt.ylabel("total value of Trades ")
plt.show()
```

- What are the companies with the highest average closing price? And what is the average closing price for these companies?

```python
grouped_data_= data.groupby(['trading_name'])['close'].mean().sort_values(ascending=False) [0:15]

plt.figure(figsize=(8, 5))
grouped_data_.plot.bar()
plt.title("Mean Closing Price by Company")
plt.xlabel("Industry")
plt.ylabel("Mean Closing Price")
plt.show()
```

- What are the companies with the lowest average closing price? And what is the average closing price for these companies?

```python
grouped_data_= data.groupby(['trading_name'])['close'].mean().sort_values(ascending=True) [0:15]

plt.figure(figsize=(8, 5))
grouped_data_.plot.bar()
plt.title("Mean Closing Price by Company")
plt.xlabel("Industry")
plt.ylabel("Mean Closing Price")
plt.show()
```
